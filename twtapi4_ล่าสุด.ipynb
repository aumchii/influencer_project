```
from datetime import datetime
import time
import requests
import sys
import tweepy
import json 
import pandas as pd
import matplotlib.pyplot as plt
import re

#อ่านไฟล์ที่รวม user 
filename ='twt_travel_total'
filename2 = filename +'.txt'
with open(filename2, 'r') as f:
    userList = f.read().splitlines()
count_index = 0 


#ไล่  user แต่ละ account ในไฟล์
for user in userList:

  pd.options.display.max_columns = 50
  pd.options.display.max_rows= 50
  pd.options.display.width= 120

  consumer_key = "" # Use your own key. To get a key https://apps.twitter.com/
  consumer_secret = ""
  access_token = ""
  access_token_secret = ""

  auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
  auth.set_access_token(access_token, access_token_secret)

  api = tweepy.API(auth)

#ส่วนของ profile 
  results = api.get_user(user)
  screen_name =results.screen_name
  location = results.location 
  user_id = results.id
  description = results.description
  name = results.name
  follower =results.followers_count
  following = results.friends_count  #following
  listed = results.listed_count
  favourites =results.favourites_count 
  statues = results.statuses_count 
  created_at = results.created_at 
  verified = results.verified
  url =results.url
  pic =results.profile_image_url
  

#เอาเข้า list ของ profile และใส่ลง dataframe
  acc_info =[] 
  acc_info.append([screen_name,name,user_id,description,follower,following,favourites,statues,created_at,pic])
  profileDf=pd.DataFrame(acc_info,columns=["screen_name","name","user_id","description","follower","following","favourites","statues","created_at","profile_pic"])
  print("profile done")

#ระบุวันที่ต้องการ 
  start_date =datetime.strptime('2020-12-01','%Y-%m-%d')
  end_date =datetime.strptime('2021-01-31','%Y-%m-%d')

#ส่วนของ tweet 
  tweet=[]
  rt_tweet=[]
  for status in tweepy.Cursor(api.user_timeline, user).items(300): 

      author=status.author.screen_name
      statusId=status.id
      text =status.text
      hashtag=status.entities['hashtags']
      retweetCount=status.retweet_count
      favoriteCount=status.favorite_count
      created_at= status.created_at
      w=10


#แยก tweet ที่เป็นการ retweet ออก
      rt_word =re.findall("^RT",text)
      if rt_word:
        rt_tweet.append([author,statusId,text,hashtag,retweetCount,favoriteCount,created_at])
        

#แยก tweet ที่ไม่ใช่วันที่ต้องการออก .ตอนนี้ไม่ได้แยกเวลา
      elif w>0:
        tweet.append([author,statusId,text,hashtag,retweetCount,favoriteCount,created_at])
        
        
#เอา tweet 1.tweet by owner 2.retweet เก็บลง dataframe        
  tweetDf =pd.DataFrame(tweet,columns=["author","statusId","text","hashtag","retweetCount","favoriteCount","created_at"])
  rt_tweetDf =pd.DataFrame(rt_tweet,columns=["author","statusId","text","hashtag","retweetCount","favoriteCount","created_at"])
  print("tweet done")

# save เป็นไฟล์นามสกุล csv  

  profileDf.to_csv("monthly_feb/profile"+'/'+filename+"/"+ user +"_Profile.csv")
  tweetDf.to_csv("monthly_feb/post"+'/'+filename+"/"+ user +"_Tweet.csv")
  rt_tweetDf.to_csv("monthly_feb/retweet"+'/'+filename+"/"+ user +"_RTTweet.csv")
  count_index = count_index+1
  print("done"+" "+ user +" "+ str(count_index))

 
```
