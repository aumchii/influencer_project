```python
import pandas as pd 
import json
import glob
from datetime import datetime 
```

```python
round= "3"
directory_list = ["data"+round+"/*.csv"]

text_files = list(map(lambda x: glob.glob(x), directory_list))
text_files = [item for sublist in text_files for item in sublist]
# print(text_files)
twt_data = [] 
twt_csv = []
for csv_file in text_files:
  q = pd.read_csv(csv_file)
  # print(csv_file)
  for index, row in q.iterrows():
    creatd_at = row['created_at'] 
    timestamp = datetime.strptime(creatd_at,'%Y-%m-%d %H:%M:%S').date()
    # print(type(timestamp))
    twt_data.append((row['author'],row['statusId'],row['text'],row['retweetCount'],row['favoriteCount'],timestamp))

twt_csv = pd.DataFrame(twt_data,columns=['author','status_Id','text','retweetCount','favoriteCount','timestamp'])
twt_csv.to_csv("travel_content"+round+".csv")
```
