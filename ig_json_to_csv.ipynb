```python
import pandas as pd 
import json
import glob
from datetime import datetime
```

```python
directory_list = ["data/*.json"]

text_files = list(map(lambda x: glob.glob(x), directory_list))
text_files = [item for sublist in text_files for item in sublist]
# print(text_files)

round_count = 0 

dataset = []

for t in text_files:
  with open (t,'r') as f: 
    data = json.load(f)
    # print(t)

  ig_content=data.popitem()
  content = ig_content[1]
  ig_name =None
  content_list = []
  text_list= []
  for item in content: 
    # likes =item['edge_media_preview_like']
    type_ = item['__typename' ] 
    # print(type_ )
    if type_ =='GraphVideo':
      try:  
        caption = item['edge_media_to_caption']['edges'][0]
      except IndexError: 
        caption ='-'
      likes = item['edge_media_preview_like']['count']
      ig_id = item['owner']['id']
      post_id =item['id']
      post_created_at = item['taken_at_timestamp']
      timestamp = str(datetime.fromtimestamp(post_created_at))
      timestamps = datetime.strptime(timestamp,'%Y-%m-%d %H:%M:%S').date()
      
    
      ig_name = item['username']
      comment_count = item['edge_media_to_comment']['count']
      vdo_view_count = item['video_view_count']
     

    else: 
      try: 
        caption = item['edge_media_to_caption']['edges'][0]
      except IndexError: 
        caption ='-'
      likes = item['edge_media_preview_like']['count']
      ig_id = item['owner']['id']
      post_id =item['id']
      post_created_at = item['taken_at_timestamp']
      timestamp = str(datetime.fromtimestamp(post_created_at))
      timestamps = datetime.strptime(timestamp,'%Y-%m-%d %H:%M:%S').date()
      
      ig_name = item['username']
      comment_count = item['edge_media_to_comment']['count']
      vdo_view_count = 0 


  
     


    


    # print('likes',likes)
    # print('ig id',ig_id)
    # print('post id',post_id)
    # print('ig name',ig_name)
    # print('comment', comment_count)
    # print('create at',post_created_at)
    # print(caption)
    text_list.append(caption)
    content_list.append([ig_name,ig_id,likes,post_id,timestamps,comment_count,vdo_view_count])

  # print("end item in profile ")

    # a= vdo[1]['text']
    # print(caption)
    # print(a)
  caption_text = []
  for item in text_list : 
    try:
      s = item['node']['text']
    except TypeError: 
      # print("TypeError")
      s = '-'
    round_count =round_count+1 
    # print(round_count)
      
    caption_text.append(s)
    # print(s)

  df_caption =pd.DataFrame(caption_text, columns=["caption"])
  df = pd.DataFrame(content_list, columns=["ig_name","ig_id","likes","post_id","timestamp","comment_count","vdo_view_count"] )
  instagram_content = pd.concat([df,df_caption], axis=1)
  # ***********
  instagram_content.to_csv("data2/"+ig_name+".csv")
 
 
```
